```python
import os

from dotenv import load_dotenv
from openai import OpenAI
```

```python
load_dotenv("./.env")
API_KEY = os.environ.get("OPENAI_API_KEY")
```

```python
client = OpenAI(api_key=API_KEY)
```

```python
response = client.responses.create(
  model="gpt-5-nano", # use 'gpt-5', 'gpt-5-mini', or 'gpt-5-nano' as model
  input="Who scored the winning goal in the Champion's League final in 2000?",
  reasoning={
      "effort": "minimal" # level of reasoning
  },
  text={
      "verbosity": "low" # i.e. output length
  }
)
```

```python
display(response.to_dict())
```

```python
print(response.output[1].content[0].text)
```
