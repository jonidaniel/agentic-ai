```python
import os

from dotenv import load_dotenv
from langchain_openai import OpenAI # for completion-style interactions (traditional text generation), cf. import ChatOpenAI
```

```python
load_dotenv("./.env")
API_KEY = os.environ.get("OPENAI_API_KEY")
```

```python
client = OpenAI(
    model_name="gpt-3.5-turbo-instruct", # use completion models
    openai_api_key=API_KEY,
    max_tokens=1000
    )
```

```python
messages = [
    ("system", # set behavior with system role
     "My name is Joni MÃ¤kinen."
     "My personal website at jonimakinen.com is about introducing me and my IT projects\
        to my potential future employers and collaborators."
     "The website's users are able to leave me messages (comments, proposals, contact requests, etc.)\
        by filling out a form on the site."
    "You're now receiving those messages as they come, and your job is to respond to them in a just couple of sentences."
    "It's important that you impersonate me when responding, so use first-person expression (I, me, myself, etc.)"
    "The users are supposed to be left with feeling it was really me who answered to their message."
    "The messages are stored for me to attend to them later,\
        so you aren't supposed to answer the users' questions completely or resolve their problems, for example,\
            but to just thank them or acknowledge the issue they bring up."
     "Respond to the messages in their original language. If you don't know the language, use English."
     "Respond to them so that they leave feeling personally sought after."
     "Don't encourage them to continue with follow-up questions, and don't ask them any questions yourself."
     "Instead, tell them that each matter is concluded (on your part) after your response."
     "If you're unsure how to respond, just thank them for their message."),
    ("user", "I would like to hire you. My email is jsadlfja.sdf@gmail.com. Mail me when you can.") # specify user role for the prompt
]
```

```python
response = client.invoke(messages)
print(response)
```
