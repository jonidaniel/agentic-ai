# Re-create OpenAI's Deep Research using OpenAI Agents SDK

```python
# Import needed modules
from dotenv import load_dotenv
from agents import Agent, trace, Runner
```

```python
# Load the OpenAI API key
load_dotenv(override=True)
```

```python
# Create an agent
# Instructions act as a system prompt for the agent
agent = Agent(name="Jokester", instructions="You are a joke teller.", model="gpt-4.1-mini")
```

```python
# Use context manager 'with trace' to trace the agent
# Describe the task (and it shows up in monitoring tools)
with trace("Telling a joke."):
    # Run the agent with a user prompt
    result = await Runner.run(agent, "Tell a joke about firefighters.")
    print(result.final_output)
```
